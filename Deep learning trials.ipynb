```python
ortho=initializers.Orthogonal()
"""

Orthogonal initialization is a weight initialization technique used in deep learning
neural networks. It involves initializing the weights of a layer with an orthogonal 
matrix. An orthogonal matrix is a square matrix whose columns and rows are 
orthogonal unit vectors, meaning that they are mutually perpendicular and 
have a length of 1.

The main idea behind orthogonal initialization is to ensure that 
the weights in the network do not explode
or vanish during the forward and backward passes, which can lead to better 
convergence during training. It can be particularly useful in recurrent neural network
s (RNNs) and transformers, where the vanishing gradient problem can be prevalent.

"""
```

```python
model1.add(Dense(4,activation='sigmoid',kernel_initializer=ortho))
```

```python
identity=initializers.Identity()
"""Identity initialization, also known as "initializing with identity" 
or "identity weight initialization," is a specific technique used in deep learning 
and neural network training. The primary purpose of identity initialization is to 
set the initial weights of a neural network layer in such a way that it encourages 
the layer to initially act as an identity function during the early stages of 
training.

In a neural network, the weights of each layer are randomly initialized before 
training begins. The idea behind identity initialization is to provide a good 
starting point for the optimization process, especially in scenarios where the 
model architecture involves shortcut connections, such as in residual networks 
(ResNets) or densely connected networks (DenseNets)."""
```

```python
model1.add(Dense(4,activation='sigmoid',kernel_initializer=identity))
```

```python
from keras import constraints
#an update to the network that checks the size of the weights, and if the size 
#exceeds a predefined limit,the weights are rescaled so that their size 
#is below the limit or between a range.
```

```python
from keras.layers import Dense,Activation,Dropout
```

```python
model=models.Sequential()
```

```python
non_neg=constraints.NonNeg()
# weight are not to be negetive
model.add(
   Dense(512, activation = 'relu', input_shape = (784,), 
   kernel_constraint= non_neg)
)
```

```python
unit_norm=constraints.UnitNorm()
# weight are not to be negetive
model.add(
   Dense(512, activation = 'relu', input_shape = (784,), 
   kernel_constraint= unit_norm)
)
```

```python
radia_const=constraints.RadialConstraint()
# weight in 2D COLV in radia
model.add(
   Dense(512, activation = 'relu', input_shape = (784,), 
   kernel_constraint= radia_const)
)
```

```python
unit_norm=constraints.MinMaxNorm(0.23,1.3)
# weight are in between max and min
model.add(
   Dense(512, activation = 'relu', input_shape = (784,), 
   kernel_constraint= unit_norm)
)
```

```python
unit_norm=constraints.MaxNorm(max_value=2)
# weight are below or equal to 2
model.add(
   Dense(512, activation = 'relu', input_shape = (784,), 
   kernel_constraint= unit_norm)
)
```

```python
l1_reg=regularizers.L1(0.1)
# l1 argument is lemda hyper prameter
l1l2_reg=regularizers.L1L2(0.1,0.1)
# l1 argument is lemda hyper prameter # l2 argument is lemda hyper prameter
l2_reg=regularizers.L2(0.1)
# l2 argument is lemda hyper prameter
orthogonal_reg=regularizers.OrthogonalRegularizer(0.1)
# developed to prevent deep CNNs from training instability and feature redundancy
```

```python
unit_norm=regularizers.L1()
# weight are not to be negetive

"kernel = weights"
"bias = biases of activation functions"
"activity = the output  of activation function"
# Yes, you're correct. The distinction you've made is accurate.
# Kernel regularizers primarily affect the weights of the neural 
# network during the backward propagation (backpropagation) phase of training,
# while activity regularizers influence the output activations of the 
# neurons during the forward propagation phase.


model.add(
   Dense(512, activation = 'relu', input_shape = (784,),kernel_regularizer=l1_reg)
)

model.add(
   Dense(512, activation = 'relu', input_shape = (784,),kernel_regularizer=l2_reg)
)
model.add(
   Dense(512, activation = 'relu', input_shape = (784,),kernel_regularizer=l1l2_reg)
)
model.add(
   Dense(512, activation = 'relu', input_shape = (784,),kernel_regularizer=orthogonal_reg)
)
```

## Loss Function in Deep Learning
## Regression
    MSE(Mean Squared Error)
    MAE(Mean Absolute Error)
    Hubber loss
## Classification

    Binary Classification Loss Function
        Binary cross-entropy

        Hinge Loss

        This type of loss is used when the target variable has 1 or -1 as class labels. It penalizes the model when there is a difference in the sign between the actual and predicted class values.
        These are particularly used in SVM models.
    
    Categorical cross-entropy
    
    If we take a dataset like Iris where we need to predict the three-class labels: Setosa, Versicolor and Virginia, in such cases where the target variable has more than two classes Multi-Class Classification Loss function is used.
## AutoEncoder
    KL Divergence
    
    Kullback Leibler Divergence Loss calculates how much a given distribution is away from the true distribution. These are used to carry out complex operations like autoencoder where there is a need to learn the dense feature representation.
## GAN
    Discriminator loss
    Minmax GAN loss
## Object detection

## Word embeddings
